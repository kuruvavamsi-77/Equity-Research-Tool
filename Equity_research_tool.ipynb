# Data-Ingestion

```python
from langchain_community.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
from langchain.document_loaders import PyPDFLoader
```

```python
# loader = WebBaseLoader("https://www.moneycontrol.com/india/stockpricequote/computers-software/tataconsultancyservices/TCS")
# loader =PyPDFLoader("/content/drive/MyDrive/211FA07077 DMT T3.pdf")
loader = UnstructuredURLLoader(["https://www.moneycontrol.com/india/stockpricequote/computers-software/tataconsultancyservices/TCS"])
data = loader.load()
data
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0,separators=["\n\n","\n"," ",""])
texts = text_splitter.split_documents(data)
```

```python
len(texts)
```

```python
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

# Small, free model (works well enough for PDF QA)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.from_documents(texts,embeddings)
db
```

# Retrivers

```python

retriver = db.as_retriever(search_type="similarity", search_kwargs={"k":2})
retriver
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
OPENROUTER_API_KEY = "
llm = ChatOpenAI(
    api_key=OPENROUTER_API_KEY,
    base_url="https://openrouter.ai/api/v1",
    model="mistralai/mistral-7b-instruct",  # or "anthropic/claude-3.5-sonnet", "gpt-4o-mini", etc.
    temperature=0
)

# qa = RetrievalQA.from_chain_type(
#     llm=llm,
#     retriever=db.as_retriever()
# )

# # 5) Ask queries on your PDF
# query = "give me the summary"
# answer = qa.run(query)
# print(answer)
```

```python
question = "give me the summary"
```

```python
retrieved_docs = retriver.invoke(question)
context = "\n\n".join(doc.page_content for doc in retrieved_docs)
```

```python

print(context)
```

```python
from langchain.prompts import PromptTemplate
prompt = PromptTemplate(
    template="""
      You are a helpful assistant.
      Answer ONLY from the provided pdf context.
      If the context is insufficient, just say you don't know.

      {context}
      Question: {question}
    """,
    input_variables = ['context', 'question']
)
```

```python
query = prompt.invoke({"context":context,"question":question})
```

```python
response = llm.invoke(query)
print("**",question,"**")
print(response.content)
```

```python
def format_docs(retrieved_docs):
  context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

**Pipeline**

```python
from langchain_core.runnables import RunnableSequence, RunnableParallel,RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parallel_chain = RunnableParallel({"context":retriver|format_docs,"question":RunnablePassthrough()})
# print(parallel_chain.invoke(question))
chain = parallel_chain|prompt|llm|parser
response = chain.invoke(question)
print(response)
```

```python
import gradio as gr
```

```python
def ask_from_url(url, question):
    try:
        # 1️⃣ Load content from the given URL
        loader = UnstructuredURLLoader([url])
        data = loader.load()

        # 2️⃣ Split text into manageable chunks
        text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200,
            separators=["\n\n", "\n", " ", ""]
        )
        texts = text_splitter.split_documents(data)

        # 3️⃣ Create embeddings and vectorstore
        embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        db = FAISS.from_documents(texts, embeddings)
        retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})

        # 4️⃣ Build the dynamic chain
        parallel_chain = RunnableParallel({
            "context": retriever | format_docs,
            "question": RunnablePassthrough()
        })

        chain = parallel_chain | prompt | llm | parser

        # 5️⃣ Get response
        response = chain.invoke(question)
        return response

    except Exception as e:
        return f"❌ Error: {str(e)}"

# ------------------ Gradio UI ------------------
with gr.Blocks() as demo:
    gr.Markdown("## 🌐 LangChain Web QA Bot")
    gr.Markdown("Paste a URL and ask any question based on that webpage.")

    url = gr.Textbox(label="Enter website URL", placeholder="https://example.com/article")
    question = gr.Textbox(label="Enter your question", placeholder="What is this article about?")
    answer = gr.Textbox(label="Answer", lines=6)

    btn = gr.Button("Ask")
    btn.click(fn=ask_from_url, inputs=[url, question], outputs=answer)

demo.launch(share=True)
```
